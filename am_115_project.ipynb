# Simulate One Twitter Session

```python
# Main helper function to simulate a single twitter session
def simulate_twitter_session(
  max_session_length_seconds,
  avg_tweets_btw_ads,
  avg_time_per_tweet,
  avg_time_per_ad,
  interaction_threshold_seconds,
  cpc,
  ad_tolerance_per_minute,
  do_print=True,
  custom_tweet_dist=None,
  return_all=False,
):
  # Initialize variables
  t = 0
  n_tweets_seen = 0
  n_ads_seen = 0
  n_ad_interactions = 0
  
  time_on_tweets = 0
  time_on_ads = 0
  
  all_times = []
  all_sequence = []
  
  ad_time_ratio = 0
  
  # Loop until user has reached session time limit
  while (t < max_session_length_seconds):
    
    # Generate random number of tweets before next ad
    n_tweets_before_ad = custom_tweet_dist(avg_tweets_btw_ads) if custom_tweet_dist is not None else np.random.poisson(avg_tweets_btw_ads)
      
    time_per_tweet= np.random.exponential(avg_time_per_tweet, size=n_tweets_before_ad)
    time_on_ad = np.random.exponential(avg_time_per_ad)
    
    time_on_tweets += time_per_tweet.sum()
    time_on_ads += time_on_ad
    
    t += time_per_tweet.sum() + time_on_ad
    n_tweets_seen += n_tweets_before_ad
    n_ads_seen += 1
    
    # Check if ad tolerance violated
    all_times.extend(list(time_per_tweet))
    all_times.append(time_on_ad)
    all_sequence.extend([0] * n_tweets_before_ad)
    all_sequence.append(1)
    
    # Calculate how many ads were seen in last 60 seconds
    last_60_seconds_idx = np.argmax(np.cumsum(np.array(all_times)[::-1]) > 60)
    last_60_seconds_n_ad = np.array(all_sequence)[-(last_60_seconds_idx + 1):].sum()
    
    if (last_60_seconds_n_ad > ad_tolerance_per_minute):
      # Final calculations before return
      ad_time_ratio = time_on_ads / t
      revenue = n_ad_interactions * cpc

      if (do_print):
        print(f"Ad tolerance exceeded: n_tweets={n_tweets_seen}, n_ads={n_ads_seen}, n_ad_interactions={n_ad_interactions}, revenue={revenue}")
        print(f"Time spent: total={t}s, tweets={time_on_tweets}s, ads={time_on_ads}s, ratio={ad_time_ratio}")
      
      if (return_all):
        return revenue, n_tweets_seen, n_ads_seen, n_ad_interactions, ad_time_ratio, True
        
      return revenue
    
    # Check if user interacted with ad
    if (time_on_ad > interaction_threshold_seconds):
      n_ad_interactions += 1
    
  # Final calculations before return
  ad_time_ratio = time_on_ads / t
  revenue = n_ad_interactions * cpc
    
  if (do_print):
    print(f"Simulation finished: n_tweets={n_tweets_seen}, n_ads={n_ads_seen}, n_ad_interactions={n_ad_interactions}, revenue={revenue}")
    print(f"Time spent: total={t}s, tweets={time_on_tweets}s, ads={time_on_ads}s, ratio={ad_time_ratio}")
  
  if (return_all):
    return revenue, n_tweets_seen, n_ads_seen, n_ad_interactions, ad_time_ratio, False
  
  return revenue
```

```python
# Run an example simulation
ctr = 0.0164
cpc = 1.35

avg_time_per_tweet = 10
avg_time_per_ad = 5
interaction_threshold_seconds = -np.log(ctr) * avg_time_per_ad # time x such that P(view_time > x) = ctr
avg_tweets_btw_ads = 5
ad_tolerance_per_minute = 3

max_session_length_minutes = 30

simulate_twitter_session(
  max_session_length_minutes * 60,
  avg_tweets_btw_ads,
  avg_time_per_tweet,
  avg_time_per_ad,
  interaction_threshold_seconds,
  cpc,
  ad_tolerance_per_minute,
)
```

# Simulate Using Empirical Mean

```python
# Do a single simulation using the empirical data to obtain session statistics
# For Table 1 in final report
ctr = 0.0164
cpc = 1.35

avg_time_per_tweet = 10
avg_time_per_ad = 5
interaction_threshold_seconds = -np.log(ctr) * avg_time_per_ad # time x such that P(view_time > x) = ctr
ad_tolerance_per_minute = 3

revenue, n_tweets_seen, n_ads_seen, n_ad_interactions, ad_time_ratio, session_ended_early = 0, 0, 0, 0, 0, 0
nsim = 1000

for i in range(nsim):
  r, t, a, i, atr, e = simulate_twitter_session(
      max_session_length_minutes * 60,
      actual_avg_tweets_btw_ads,
      avg_time_per_tweet,
      avg_time_per_ad,
      interaction_threshold_seconds,
      cpc,
      ad_tolerance_per_minute,
      do_print=False,
      return_all=True,
    )
  
  revenue += r
  n_tweets_seen += t
  n_ads_seen += a
  n_ad_interactions += i
  ad_time_ratio += atr
  session_ended_early += e
```

```python
# Calculate simulation averages and print
revenue /= nsim
n_tweets_seen /= nsim
n_ads_seen /= nsim
n_ad_interactions /= nsim
ad_time_ratio /= nsim
session_ended_early /= nsim

print(f"Revenue={revenue}, session ended early %={session_ended_early}")
print(f"Avg simulation results: n_tweets_seen={n_tweets_seen}, n_ads_seen={n_ads_seen}, n_ad_interactions={n_ad_interactions}, ad_time_ratio={ad_time_ratio}")
```

```python
# Simulate using Poisson distribution across a variety of ad tolerances
nsim = 5000
ad_tolerances = [1, 2, 3, 4, 5, 6, 7]
revs_by_ad_tolerance_poisson = []

for ad_tolerance_per_minute in ad_tolerances:
  tot_revenue = 0

  for i in range(nsim):
    tot_revenue += simulate_twitter_session(
      max_session_length_minutes * 60,
      actual_avg_tweets_btw_ads,
      avg_time_per_tweet,
      avg_time_per_ad,
      interaction_threshold_seconds,
      cpc,
      ad_tolerance_per_minute,
      do_print=False,
    )

  revs_by_ad_tolerance_poisson.append(tot_revenue / nsim)
```

```python
# Simulate using empirical distribution across a variety of ad tolerances
nsim = 5000
ad_tolerances = [1, 2, 3, 4, 5, 6, 7]
revs_by_ad_tolerance_twitter = []

for ad_tolerance_per_minute in ad_tolerances:
  tot_revenue = 0

  for i in range(nsim):
    tot_revenue += simulate_twitter_session(
      max_session_length_minutes * 60,
      avg_tweets_btw_ads,
      avg_time_per_tweet,
      avg_time_per_ad,
      interaction_threshold_seconds,
      cpc,
      ad_tolerance_per_minute,
      do_print=False,
      custom_tweet_dist=data_tweet_nmbr,
    )

  revs_by_ad_tolerance_twitter.append(tot_revenue / nsim)
```
